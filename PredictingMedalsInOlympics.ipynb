```python
train = teams[teams['year']<2012].copy()
test = teams[teams['year']>=2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
inputs = ['athletes','prev_medals','events']
target = 'medals'
model.fit(train[inputs],train[target])
prediction = model.predict(test[inputs])
```

```python
test['prediction']=prediction
```

```python
test
```

```python
test.loc[test['prediction']<0,'prediction']=0
```

```python
test
```

```python
test['prediction']=test['prediction'].round()
test
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
error = mean_absolute_error(test['medals'],test['prediction'])
```

```python
errors = (test['medals']-test['prediction']).abs()
errors
```

```python
error_by_team = errors.groupby(test['team']).mean()
error_by_team
```

```python
medals_by_team= test['medals'].groupby(test['team']).mean()
error_ratio = error_by_team/medals_by_team
error_ratio
```
